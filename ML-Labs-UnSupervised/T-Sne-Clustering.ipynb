```python
# T-SNE CLUSTERING
# T-Distributed Stochastic Neighbor Embedding

"""
It maps high-dimensional space into a two or three-dimensional space which can then be visualized. 
Specifically, it models each high-dimensional object by a two- or three-dimensional point in 
such a way that similar objects are modeled by nearby points and dissimilar objects are modeled 
by distant points with high probability.
"""
# Importing Modules
from sklearn import datasets
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd
```

```python
# Loading dataset
#iris_df = datasets.load_iris()
iris =pd.read_csv('IRIS.csv')
```

```python
array=iris.values
#print(array)
X=array[:,:4]
# Defining Model
model = TSNE(learning_rate=100)
```

```python
# Fitting Model
transformed = model.fit_transform(X)
```

```python
col=[]
for i in range(150):
    if i>=0 and i<50:
        col.append('r')
    if i>=50 and i<100:
        col.append('g')
    if i>=100 and i<150:
        col.append('b')
print(col)
```

```python
# Plotting 2d t-Sne
x_axis = transformed[:, 0]
y_axis = transformed[:, 1]

plt.scatter(x_axis, y_axis, c=col)
plt.show()
```

```python
X_pca = PCA().fit_transform(X)

plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(transformed[:, 0], transformed[:, 1], c=col)
plt.subplot(122)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=col)
plt.show()
```
